```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm 
from astropy.io import fits 
from matplotlib import colors
from matplotlib.colors import LogNorm
```

```python
hdu_list = fits.open('/users/ava/downloads/2021-08-15_wholeFoV.fits')
image_data = hdu_list[0].data
hdu_list.close()
plt.imshow(np.flipud(image_data), extent=[-4,4, -4,4], zorder=0, cmap='gist_gray', norm=LogNorm(), clim=(1200.0, 18000.), alpha=0.2)
```

```python
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.xlabel('offset from Sgr A* -RA [arcsec]', size=16, labelpad=20)
plt.ylabel('offset from Sgr A* - Dec [arcsec]', size=16, labelpad=20)
plt.plot(1, 0, marker="o", markersize=20, markeredgecolor="red", markerfacecolor="green")
```
